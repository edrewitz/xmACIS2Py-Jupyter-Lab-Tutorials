# Multi-Station Data Retrieval

In this example, we will retrieve multi-station data for the past 30 days from a list of stations:['KLAX', 'KONT', 'KRAL', 'KRIV']

We will then create CSV files of this data and then open that data in Pandas.

## Imports

```python
import pandas as pd
from xmacis2py import get_multi_station_acis_data
```

```python
# Define our list of station IDs
stations = ['KLAX', 'KONT', 'KRAL', 'KRIV']
```

## Retrieving The Data With xmACIS2Py

In this example, we want to create CSV files for each station.

To do this we must change our optional argument of `to_csv=True` and in this case it is good practice to set `return_pandas_df=False`

The default path for the CSV files to save is f:XMACIS DATA/{files}

```python
get_multi_station_acis_data(['KLAX', 'KONT', 'KRAL', 'KRIV'],
                      to_csv=True,
                      return_pandas_df=False)
```
