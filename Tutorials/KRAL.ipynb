```python
'''
In this example we will make a 30-Day and 90-Day temperature and precipitation summaries for the Riverside Regional Airport in Southern California (KRAL)

We will need to import both our plot_temperature_summary and plot_precipitation_summary functions from xmacis2py
'''
# Imports
from xmacis2py import plot_temperature_summary, plot_precipitation_summary
```

```python
'''
Here we make out plots of the temperature and precipitation summaries. In this example, we aren't using a custom date range and we aren't changing the defaults to whether
or now we want to plot running means and/or running sums so we don't need to make any changes to the optional arguments. 
The function has 2 required arguments which are (station, product_type), here is what they mean:

1) station (String) - The xmACIS2 station ID as a string. 
2) product_type (String or Integer) - The type of summary. The user can choose either a 7-Day/10-Day/15-Day/30-Day/60-Day/90-Day Summary. 

You can enter the product_type as a string for example: 'Past 7 Days' or 'Past 10 Days' or 'Past 15 Days' or 'Past 30 Days' or 'Past 60 Days' or 'Past 90 Days'
To enter the product_type as an integer use a 7 for Past 7 Days, a 10 for Past 10 Days....90 for Past 90 Days. 

Lastly the functions will create two main directories. One directory (ACIS Data) will be created to house the data which is in a .CSV format. 
The other folder (ACIS Graphics) will hold the graphics created. 

'''
plot_temperature_summary('KRAL', 30)
plot_precipitation_summary('KRAL', 30)
plot_temperature_summary('KRAL', 90)
plot_precipitation_summary('KRAL', 90)
```
