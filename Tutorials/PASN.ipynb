```python
'''
In this example we will make a summary of temperature and precipitation for Saint Paul, AK (PASN) from Nov. 1st - Nov. 30th 2024

We will need to import both our plot_temperature_summary and plot_precipitation_summary functions from xmacis2py
'''
# Imports
from xmacis2py import plot_temperature_summary, plot_precipitation_summary
```

```python
'''
Here we make out plots of the temperature and precipitation summaries. In this example, we ARE using a custom date range so we NEED to make changes to the optional arguments. 
The function has 2 required arguments which are (station, product_type), here is what they mean:

1) station (String) - The xmACIS2 station ID as a string. 
2) product_type (String or Integer) - The type of summary. The user can choose either a 7-Day/10-Day/15-Day/30-Day/60-Day/90-Day Summary. 

In this case, we must enter 'Custom' or 'custom' for our product_type

The function has 2 optional arguments which are (start_date, end_date), here is what they mean:

First of all, the default values for both start_date and end_date is None. 
When using xmACIS2Py to analyze a custom date range, we need to change these None values to a start and end date in the form of a string following the following format:

YYYY-MM-DD

Thus, in this example we will use '2024-11-01' for start_date and '2024-11-30' for end_date

Lastly the functions will create two main directories. One directory (ACIS Data) will be created to house the data which is in a .CSV format. 
The other folder (ACIS Graphics) will hold the graphics created. 

'''
plot_temperature_summary('PASN', 'Custom', start_date='2024-11-01', end_date='2024-11-30')
plot_precipitation_summary('PASN', 'custom', start_date='2024-11-01', end_date='2024-11-30')
```
