# Importing xmACIS2Py

```python
from xmacis2py import(
    temperature,
    precipitation)
```

# Comprehensive Temperature Summary for the past 30 days at KRAL

```python
temperature.plot_comprehensive_summary('kral',
                                      cooling_degree_days=False)
```

View the graphic [here](https://github.com/edrewitz/xmACIS2Py/blob/main/Graphical%20Summary%20Examples/KRAL%20Comprehensive%2030%20Day%20Summary.png)

# Maximum Temperature Summary for July 2025 at KJFK.

```python
temperature.plot_maximum_temperature_summary('kjfk',
                                             product_type='Maximum Temperature Summary July 2025',
                                             start_date='2025-07-01',
                                             end_date='2025-07-31')
```

View the graphic [here](https://github.com/edrewitz/xmACIS2Py/blob/main/Graphical%20Summary%20Examples/KJFK%20Maximum%20Temperature%20Summary%20July%202025.png)

View the stats table [here](https://github.com/edrewitz/xmACIS2Py/blob/main/Graphical%20Summary%20Examples/KJFK%20Stats%20Table.png)

# Average Temperature Summary Detrended (Linear) from January 1st 1991 - December 31st 2020 at PASN.

```python
temperature.plot_average_temperature_summary('pasn',
                                             product_type='30 Year Detrended Average Temperature PASN',
                                             start_date='1991-01-01',
                                             end_date='2020-12-31',
                                             x_axis_day_interval=365,
                                             x_axis_date_format='%Y',
                                             detrend_series=True,
                                             plot_type='line',
                                             show_running_mean=False)
                                             
```

View the graphic [here](https://github.com/edrewitz/xmACIS2Py/blob/main/Graphical%20Summary%20Examples/PASN%2030%20Year%20Detrended%20Average%20Temperature%20PASN.png)

View the stats table [here](https://github.com/edrewitz/xmACIS2Py/blob/main/Graphical%20Summary%20Examples/PASN%20Stats%20Table.png)
