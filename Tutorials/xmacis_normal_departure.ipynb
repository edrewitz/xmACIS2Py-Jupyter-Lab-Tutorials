# Downloading 30-Year Climate Normals and Departures From Normal

In this example, we will download the daily, monthly and yearly normals for Denver, CO (KDEN)

Here is the documentation for the functions that download historical ACIS2 Data:

1) [Get Single Station Data](https://github.com/edrewitz/xmACIS2Py/blob/main/Documentation/xmACIS2.0/data_access.md#get_single_station_acis_data)
2) [Get Multi Station Data](https://github.com/edrewitz/xmACIS2Py/blob/main/Documentation/xmACIS2.0/data_access.md#get_multi_station_acis_data)
3) [Get Single Station Climate Normals](https://github.com/edrewitz/xmACIS2Py/blob/main/Documentation/xmACIS2.0/data_access.md#get_single_station_climate_normals)
4) [Get Multi Station Climate Normals](https://github.com/edrewitz/xmACIS2Py/blob/main/Documentation/xmACIS2.0/data_access.md#get_multi_station_climate_normals)
5) [Get Single Station Departures From Normal](https://github.com/edrewitz/xmACIS2Py/blob/main/Documentation/xmACIS2.0/data_access.md#get_single_station_departures)
6) [Get Multi Station Departures From Normal](https://github.com/edrewitz/xmACIS2Py/blob/main/Documentation/xmACIS2.0/data_access.md#get_multi_station_departures)

## Imports

```python
import xmacis2py as xmacis
```

## Download Daily Monthly and Yearly Normals

```python
daily_normals = xmacis.get_single_station_climate_normals('kden', 
                                                          start_date='2025-01-01', 
                                                          end_date='2025-12-31')
```

```python
daily_normals
```

```python
monthly_normals = xmacis.get_single_station_climate_normals('kden', 
                                                            start_date='2025-01-01', 
                                                            end_date='2025-12-31',
                                                            interval='monthly')
```

```python
monthly_normals
```

```python
annual_normals = xmacis.get_single_station_climate_normals('kden', 
                                                            start_date='2025-01-01', 
                                                            end_date='2025-12-31',
                                                            interval='yearly')
```

```python
annual_normals
```

## Download Daily Monthly and Yearly Departures From Normal

```python
daily_departures = xmacis.get_single_station_departures('kden', 
                                                        start_date='2025-01-01', 
                                                        end_date='2025-12-31')
```

```python
daily_departures
```

```python
monthly_departures = xmacis.get_single_station_departures('kden', 
                                                        start_date='2025-01-01', 
                                                        end_date='2025-12-31',
                                                        interval='monthly')
```

```python
monthly_departures
```

```python
annual_departures = xmacis.get_single_station_departures('kden', 
                                                        start_date='2025-01-01', 
                                                        end_date='2025-12-31',
                                                        interval='yearly')
```

```python
annual_departures
```
