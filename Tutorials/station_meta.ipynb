# Station Meta Data

In these examples, we will download the station meta-data for a single station and a list of stations

## Imports

```python
import pandas as pd
import xmacis2py.data_access as acis_data
```

```python
# We will use KLAX as our station
station = 'klax'
meta = acis_data.single_station_meta(station)
```

```python
meta
```

```python
# We will now retrieve a list of station meta-data for a user-defined list of station IDs.
stations = ['KLAX', 'KJFK', 'KDEN', 'KISP']
meta = acis_data.multi_station_meta(stations)
```

```python
meta
```

```python
# Now Let's try this again and make CSVs for each station
# Default path is f:XMACIS META/{files}
acis_data.multi_station_meta(stations,
                            to_csv=True,
                            return_pandas_df=False)
```

```python
# Now let's open the CSVs for each station's meta-data
klax = pd.read_csv(f"XMACIS META/KLAX.csv")
```
